```python
import re


rating = []
for i in soup.find_all("div", class_="c066246e13 d8aec464ca"):
    element = i.find("div", class_="a3b8729ab1 d86cee9b25")
    if element:
        text = element.text
        regex = re.findall(r"\s\d+?\.?\d\s", text)
        if regex:
            rating.append(regex[0].strip())  # Add strip() to remove any surrounding whitespace
        else:
            rating.append(np.nan)
    else:
        rating.append(np.nan)

print(rating)
```

```python
cancellation = []  # Initialize an empty list to store results
count = 0  # Initialize count for tracking total divs processed

# Iterate through all div elements with the specified class
for i in soup.find_all("div", class_="c066246e13 d8aec464ca"):
    count += 1
    
    # Find the inner div with the specific class inside the current div
    inner_div = i.find("div", class_="c72df67c95 db70416f32")
    
    # Check if the inner div was found before accessing its text
    if inner_div:
        cancellation.append(inner_div.text.strip())  # Append the cleaned text to the list
    else:
        cancellation.append("No Free cancellation")  # Append default message if not found

# Print the cancellation data and the total number of divs processed

print(cancellation)
print(f"Total divs processed: {count}")
```

```python

distance = []
for i in soup.find_all("div", class_="c066246e13 d8aec464ca"):
    span = i.find("span", attrs={"data-testid": "distance"})
    if span:
        text = span.text
        regex = re.findall(r"(\d?\.?\d)", text)
        if regex:
            distance.append(regex[0].strip())  # Add strip() to remove any surrounding whitespace
        else:
            distance.append(np.nan)
    else:
        distance.append(np.nan)

print(distance)
```

```python

Breakfast = []

# Iterate through all div elements with the specified class
for i in soup.find_all("div", class_="c066246e13 d8aec464ca"):
    # Try to find the <span> element with the specified class
    span = i.find("span", class_="a19404c4d7")
    
    # Check if the span element was found
    if span:
        Breakfast.append(span.text)  # Append the text content of the span
    else:
        Breakfast.append("Not included")  # Append "nan" if the span is not found

# Print the list of breakfast items (or nan where not found)
print(Breakfast)
print(len(Breakfast))
```
